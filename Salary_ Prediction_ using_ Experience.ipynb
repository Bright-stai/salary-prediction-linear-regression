```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
x_train = np.array([2,5,3,6,7])
y_train = np.array([400,1600,800,2000,2400])
w = 0
b = 0
```

```python
x = x_train/10
```

```python
# prdiction Function
def predict(x,w,b):
    return w * x + b
```

```python
#Model function
def compute_model(x,w,b):
    
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
#Cost function
def compute_cost_function(x,y,w,b):
    m = x.shape[0]
    sum_cost = 0
    for i in range(m):
        f_wb = w * x[i] + b
        total_cost = (f_wb - y[i])**2
        sum_cost = sum_cost + total_cost
        final_cost = (1/(2*m)) * sum_cost
    return final_cost
```

```python
# Gradient descent
def compute_gradient(x,y,w,b,):
    m = x.shape[0]
    dj_dw = 0
    dj_db = 0
    for i in range(m):
        f_wb = w * x[i] + b
        dj_dw_i = (f_wb - y[i]) * x[i]
        dj_db_i = f_wb  - y[i]
        dj_dw += dj_dw_i
        dj_db += dj_db_i
    dj_dw = dj_dw / m
    dj_db = dj_db / m
    return dj_dw, dj_db
        
        
```

```python
# TRAINING
alpha = 0.1

for i in range(5000):
    dj_dw, dj_db = compute_gradient(x, y_train, w, b)
    w = w - alpha * dj_dw
    b = b - alpha * dj_db
print("final w", w)
print("final b", b)
```

```python
y_pred = predict(x,w,b)
print(y_pred)
print(y_train)
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train,y_pred, color="red")
plt.title("Salary vs Experiencee (linear Regression)")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show
```
